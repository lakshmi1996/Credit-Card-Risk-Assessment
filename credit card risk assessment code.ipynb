```python
x = credit_ra.drop(['default.payment.next.month'],axis=1)
```

```python
y = credit_ra['default.payment.next.month']
```

```python
y.head()
```

```python
#Hyperparameters optimization
params={
  
     "min_child_weight" :[1,3,4,5,6,9]
     , "max_depth":[3,4,5,6,7,10]
     ,"gamma" :[0.1,0.2,0.3,0.4]
     , "learning_rate":[0.05,0.10,0.15,0.20,0.25]
     ,"colsample_bytree":[0.1,0.2,0.3,0.5,0.8,1.0]
    
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost
```

```python
final_classifier = xgboost.XGBClassifier()
```

```python
final_classifier
```

```python
random_search = RandomizedSearchCV(final_classifier,param_distributions=params,n_iter=5,scoring='roc_auc',refit =True, verbose =3,n_jobs= -1,cv=5)
```

```python
random_search.estimator
```

```python
final_classifier = xgboost.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=1, gamma=0,
              learning_rate=0.1, max_delta_step=0, max_depth=3,
              min_child_weight=1, missing=None, n_estimators=100, n_jobs=1,
              nthread=None, objective='binary:logistic', random_state=0,
              reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
              silent=None, subsample=1, verbosity=1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score= cross_val_score(final_classifier,x,y,cv=20)
```
